```python
import pandas as pd
train = pd.read_csv('train.csv')
train
```

```python
features = list(train.columns)
cat_columns = list(train[features].select_dtypes(include=['bool','object']).columns)
cat_columns
```

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Assume: 
# df = your full dataset (e.g., train)
# cat_columns = list of your categorical column names
# 'loan_default' = your target column (categorical, binary: 0 or 1)

# Step 1: Encode categorical features numerically
X = train[cat_columns].apply(LabelEncoder().fit_transform)

# Step 2: Encode the target if it's categorical
y = LabelEncoder().fit_transform(train['loan_defaulted'])

# Step 3: Apply Chi-Square test
chi_scores = chi2(X, y)

# Step 4: Create DataFrame with scores and p-values
chi2_df = pd.DataFrame({
    'Feature': cat_columns,
    'Chi2 Score': chi_scores[0],
    'p-value': chi_scores[1]
})

# Step 5: Sort by Chi2 Score (higher = more relevant)
chi2_df.sort_values(by='Chi2 Score', ascending=False, inplace=True)

# View result
print(chi2_df)
```
